##**Mapas de calor**

Se decidió realizar un mapa de calor donde se pueda observar qué variables tienen una mayor correlación entre sí.  

La correlación es un tipo de asociación entre dos variables numéricas, que evalúa la tendencia en los datos. Se puede decir que dos variables están asociadas cuando una de ellas nos da información acerca de la otra. Por el contrario, cuando no existe asociación, el aumento o disminución de una variable no suele decir nada acerca del comportamiento de la  otra variable. Dentro del mapa de calor, los colores más cercanos al azul (1.0) nos dicen que las variables tienen relación una con la otra y los colores cercanos al amarillo (0.0) nos dice que no tiene correlación una con la otra. 


### **Devoluciones**

```python
#Ajustamos el mapa de calor de los coeficientes de correlación 
Heat_Map= sns.heatmap(abs(devoluciones.corr()), cmap = 'YlGnBu', annot=True, fmt=".2f")  
Heat_Map
```

Dentro de el dataframe devoluciones, las variables númericas no tienen una gran intensidad de relación una con la otra. Dando un resultado de 0.15 esto nos dice que la relación lineal es débil.

### **Facturacion**

```python
#Ajustamos el mapa de calor de los coeficientes de correlación 
Heat_Map= sns.heatmap(abs(facturacion.corr()), cmap = 'YlGnBu', annot=True, fmt=".2f")  
Heat_Map
```

Hablando de las variables del dataframe de facturación, se observa una correlación de 0.23, lo que nos dice que no existe relación líneal entre las variables.

###**Gastos y Costos**

```python
fig = px.box(gastos_23, y='Importe')
fig.update_layout(title="Importe")
fig.update_layout(xaxis_title="Importe", yaxis_title="")
fig.update_traces(marker_color='darkred')
fig.show()
```

```python
fig =  px.histogram(gastos_23, x='Importe')
fig.update_layout(title=" Histograma Importe")
fig.update_layout(xaxis_title="Importe", yaxis_title="Count")
fig.update_traces(histnorm="percent")
fig.update_traces(marker_color='darkred')
fig.show()
```

```python
fig =  px.histogram(gastos_23, x='TOTAL MX')
fig.update_layout(title="Histograma Total MX")
fig.update_traces(histnorm="percent")
fig.update_traces(marker_color='darkred')
fig.show()
```

```python
fig = px.box(gastos_23, x='TIPO GASTO', y='Importe')
fig.update_layout(title="Importe por cada Tipo de Gasto")
fig.update_layout(xaxis_title="Tipo de Gasto", yaxis_title="Importe")
fig.update_traces(marker_color='darkred')
fig.update_traces(marker=dict(size=4))
fig.show()
```
